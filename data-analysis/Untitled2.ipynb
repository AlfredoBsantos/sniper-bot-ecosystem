```python
import pandas as pd

# Define o nome do arquivo que você baixou
# (Renomeie aqui se você deu um nome diferente ao seu arquivo)
file_path = 'proxy_transactions.csv' 

try:
    # Carrega o CSV para uma tabela (DataFrame) do pandas
    df = pd.read_csv(file_path)
    
    print(f"Dataset do Proxy carregado com sucesso!")
    print(f"Total de {len(df)} transações históricas para análise.")
    print("-------------------------------------------")
    
    # ---------------- Análise Rápida ----------------
    
    # 1. Quais são as interações mais comuns?
    print("\n--- Frequência dos Métodos Chamados ---")
    print(df['Method'].value_counts().head(10))
    
    # 2. Quem são os maiores 'jogadores'?
    print("\n--- Top 10 Endereços que Mais Interagiram ---")
    print(df['From'].value_counts().head(10))
    
    # 3. Vamos ver as 5 primeiras linhas para ter uma ideia dos dados
    print("\n--- Amostra dos Dados (Primeiras 5 Linhas) ---")
    display(df.head())
    
except FileNotFoundError:
    print(f"--- ERRO: Arquivo '{file_path}' não encontrado! ---")
    print("Por favor, verifique se o arquivo CSV está na mesma pasta que este notebook.")
```
